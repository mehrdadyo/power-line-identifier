```python
import torch
from torch.utils.data import DataLoader

import numpy as np

from useful_things.dataloader import *
from useful_things.networks import *
from useful_things.helper_functions import *
```

```python
N = {}
N['noUtil'] = 236
N['noVeg'] = 99
N['Veg'] = 90

partition = {'all':[]}
for i in range(N['noUtil']):
    partition['all'].append((0,i))
for i in range(N['noVeg']):
    partition['all'].append((1,i))
for i in range(N['Veg']):
    partition['all'].append((1,i))
    
params = {'batch_size': 64,
         'shuffle': True,
         'num_workers': 8}

data_generator = data.DataLoader(UtilityDataset(partition['all']), **params)
```

```python
model = torch.load('best_model')

use_cuda = torch.cuda.is_available()
device = torch.device("cuda:0" if use_cuda else "cpu")

output = {'noUtil':[],'noVeg':[],'Veg':[]}

model.eval()
model.cuda()
for imgs, labels, lat, long, cat in data_generator:
    imgs, labels = imgs.to(device), labels.to(device)
    scores = model(imgs)
    pred_class = scores.argmax(1)
    
    for i in range(labels.shape[0]):   
        if pred_class[i] == 0:
            output['noUtil'].append((lat[i],long[i]))
        elif pred_class[i] ==1:
            output['noVeg'].append((lat[i],long[i]))
        else:
            output['Veg'].append((lat[i],long[i]))
```
